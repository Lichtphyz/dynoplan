```python
df_alg = df.groupby('alg').mean(numeric_only=True).reset_index()
df_alg.round(2)[['alg','success_rate','time_first_solution','cost_first_solution','cost_at_60','last_cost']]
```

```python
df_alg_m = df_alg.melt(id_vars=['alg', 'success_rate','time_first_solution'],  # columns to keep as-is
        var_name='cost_type',                  # new column for the old column names
        value_name='cost')                   # new column for the values
df_alg_m['cost_type'] = df_alg_m['cost_type'].str.replace("cost_at_","")
df_alg_m = df_alg_m[df_alg_m['alg']!='RRT']
df_alg_m = df_alg_m[~df_alg_m['cost_type'].str.contains("length")]
df_alg_m
```

```python
ax = sns.lineplot(data=df_alg_m, x='cost_type', y='cost', hue='alg')
ax.set_title('Mean Perforamnce Over 5 Env (when solvable)');
```

```python
fig, ax = plt.subplots(figsize=(6, 3))  # (width, height) in inches — adjust to taste
sns.lineplot(data=df_alg_m, x='cost_type', y='cost', hue='alg',
             palette=['orange', 'green'], ax=ax)
ax.set_title('Mean Cost(s) Over 5 Env (when solvable)')
```
